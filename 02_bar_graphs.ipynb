```python
#offense, defense, and special teams groups
offense_pos = ["QB", "RB", "FB", "WR", "TE", "OT", "OG", "C", "OL"]
defense_pos = ["DE", "DT", "DE/DT", "LB", "CB", "S", "DB"]
sp_teams_pos = ["K", "P"]

arrest_data = pd.read_csv('Resources/ArrestIncidents.csv')
arrest_data["YEAR"] = arrest_data["DATE"].str[-4:]

arrest_data.head()
```

```python
team_count = arrest_data["TEAM"].value_counts()
team_count.head()
```

```python
position_count = arrest_data["POSITION"].value_counts()
position_count
```

```python
import matplotlib.pyplot as plt


# Create a bar graph
plt.figure(figsize=(10, 6))
bars = position_count.plot(kind='bar', color='purple')

# Add labels and title
plt.xlabel('Position')
plt.ylabel('Arrests')
plt.title('Arrests by Position')

# Add text annotations on top of each bar
for bar, count in zip(bars.patches, position_count.values):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), str(count),
             ha='center', va='bottom', color='black')

# Show the graph
plt.show()

```

```python
import matplotlib.pyplot as plt


categories = ['DUI', 'Drugs', 'Domestic violence', 'Assault', 'Disorderly conduct',
              'Stalking', 'DUI, assault', 'License, guns', 'Guns, drugs, reckless driving', 'Murder']
values = [116, 52, 47, 38, 19, 1, 1, 1, 1, 1]

# Create a bar graph
plt.figure(figsize=(12, 6))
bars = plt.bar(categories, values, color='purple')

# Add labels and title
plt.xlabel('Crime')
plt.ylabel('Arrests')
plt.title('Defense Crime and Arrests')

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45, ha='right')

# Display count values on top of the bars
for bar, value in zip(bars, values):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), str(value),
             ha='center', va='bottom', fontsize=8, color='black')

# Show the graph
plt.tight_layout()
plt.show()
```
